```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\chast\Documents\BackupDocuments\Anime.csv")

df
```

```python
standings_df = df[["Name","K-Popularity","K-rated","Mal-Popularity", "Mal-rated"]]
standings_df
```

```python
new_standings_df = standings_df.set_index("Name")
new_standings_df
```

```python
plt.plot(new_standings_df.index,standings_df["K-Popularity"], label="Kitsu Popularity")
plt.plot(new_standings_df.index,standings_df["Mal-Popularity"], label="MyAnimeList Popularity")
plt.xticks(ticks=np.arange(30),labels=standings_df['Name'])
plt.legend()
plt.title("Popularity on MyAnimeList vs Kitsu")
plt.xlabel("Animes")
plt.ylabel("Popularity")
plt.grid()

plt.show()
```

```python
plt.plot(new_standings_df.index,standings_df["K-rated"], label="Kitsu Ratings")
plt.plot(new_standings_df.index,standings_df["Mal-rated"], label="MyAnimeList Ratings")
plt.xticks(ticks=np.arange(30),labels=standings_df['Name'])
plt.legend()
plt.title("Viewer Ratings on MyAnimeList vs Kitsu")
plt.xlabel("Animes")
plt.ylabel("Ratings")
plt.grid()

plt.show()
```

```python
animes_on_platform_df = df.groupby(['Original Platform'],as_index=False)['Name'].count()
animes_on_platform_df
```

```python
animes_on_platform_df.plot.bar(x='Original Platform', y='Name')
plt.legend(loc="best")
plt.title("Amount of Animes Originally Released on Each Viewing Network")
plt.xlabel("Network")
plt.ylabel("Name")
plt.grid()

plt.show()
```

```python
language_df = df.groupby(['Original Style'],as_index=False)['Name'].count()
language_df
```

```python
styles = ["English","English Dub","Japanese/Sub"]
count = [4, 6, 20]

plt.pie(count, labels=styles, startangle=100)
plt.show()
```

```python
animes_per_year_df = df.groupby(['Release Year'],as_index=False)['Name'].count()
animes_per_year_df
```

```python
animes_per_year_df.plot.bar(x='Release Year', y='Name')
plt.xticks(ticks=np.arange(16),labels=animes_per_year_df['Release Year'])
plt.legend()
plt.title("Amount of Animes Produced Each Year")
plt.xlabel("Years")
plt.ylabel("Amount of Animes")
plt.grid()

plt.show()
```
