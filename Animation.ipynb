```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\chast\Documents\BackupDocuments\Animation.csv")

df
```

```python
ratings_df = df[["Name", "IMDB rating", "rotten tomatoes rating", "Google Rating "]]
ratings_df
```

```python
df.columns
```

```python
shows_per_year_df = df.groupby(['Release Year'],as_index=False)['Name'].count()
shows_per_year_df
```

```python
shows_per_network_df = df.groupby(['Original Network'],as_index=False)['Name'].count()
shows_per_network_df
```

```python
shows_per_network_df.plot.bar(x='Original Network', y='Name')
plt.legend(loc="best")
plt.title("Amount of Shows On Each Network")
plt.xlabel("Network")
plt.ylabel("Show name")
plt.grid()

plt.show()
```

```python
episodes_per_network_df = df.groupby(['Original Network'],as_index=False)['Number of Episodes'].sum()
episodes_per_network_df
```

```python
episodes_per_network_df.plot(x='Original Network', y='Number of Episodes',figsize=(20,20))
plt.xticks(ticks=np.arange(11),labels=episodes_per_network_df['Original Network'])
plt.legend()
plt.title("Amount of Episodes Each Network Produced")
plt.xlabel("Networks")
plt.ylabel("Amount of Episodes")
plt.grid()

plt.show()
```
